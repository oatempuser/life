```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import matplotlib as mpl
import time
import threading
from ipywidgets import interact, widgets
from random import randrange

from random import choice
import sys

space_dim = (30,60)
energy_dim = 100
gen_types = [1,2,3,4]
```

```python
class Gen:
    def __init__(self, gen_type = choice(gen_types)):
        self.gen_type = gen_type
    def __str__(self):
        return str(self.gen_type)
    def action(self, target):
        return self

class Genom:
    def __init__(self, gen = Gen()):
        self.body = [gen]
    def __str__(self):
        out = ''
        for i in self.body:
            out += str(i)
    
    def append(self, gen):
        self.body.append(gen)
        return out  
```

```python
class Enemy:
    def __init__(self, energy, genom = Genom()):
        self.genom = genom
        self.energy = energy
        self.addres = []
    def __str__(self):
        return str(self.energy)
    
    def get_addres(self):
        self.addres.append([randrange(space_dim[0]),randrange(space_dim[1])])
        print (self.addres)
   
    def tick(self):
        Enemy.get_addres(self)
        if self.energy < 1:
            del self
            return
        #self.energy -= 1

```

```python
class Population:
    def __init__(self, space_dim = space_dim):
        self.enemys = np.empty(space_dim, dtype = Enemy)
        self.aval_space = np.transpose(np.where(self.enemys == None))
        self.len_aval_space = len(self.aval_space)
        self.count = 0
        
    def __str__(self):
        return str(self.count)
    
    def new_random_enemy(self):
        if (self.len_aval_space < 1): return
        self.enemys[tuple(choice(self.aval_space))] = Enemy(energy = randrange(100))#energy = randrange(100))        
        self.aval_space = np.transpose(np.where(self.enemys == None))
        self.len_aval_space = len(self.aval_space)
        self.count += 1
    
    def tick(self):
        for i in np.transpose(np.where(self.enemys != None)):
            e = self.enemys[tuple(i)]
            e.tick()
            if (e.energy < 1):
                self.enemys[tuple(i)] = None
                self.aval_space = np.transpose(np.where(self.enemys == None))
                self.len_aval_space = len(self.aval_space)
                self.count -= 1
                continue
            
        Population.new_random_enemy(self)
  
    
    def view_space(self):
        df = pd.DataFrame(data=self.enemys)
        df = df.replace([None], 0)
        df = df.style.set_properties(**{'font-size': '10px'}) 
        df = df.set_properties(**{'background-color': 'black',
                           'color': 'lawngreen',
                           'border-color': 'white'})
        return df
        
```
